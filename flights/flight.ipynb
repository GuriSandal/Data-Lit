```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/flights.csv")
df.head()
```

```python
df.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
df.head()
```

```python
ax = df.plot.hist(bins=int(180/5), edgecolor='black')
ax.set_title('Histogram of Arrival Delays')
ax.set_xlabel('Delay (min)')
ax.set_ylabel('Flights')
```

```python
airlines = ['United Air Lines Inc.', 'JetBlue Airways', 'ExpressJet Airlines Inc.',
         'Delta Air Lines Inc.', 'American Airlines Inc.']

# Iterate through the five airlines
for airline in airlines:
    # Subset to the airline
    subset = df[df['name'] == airline]
    
    # Draw the density plot
    sns.distplot(subset['arr_delay'], hist = False, kde = True,
                 kde_kws = {'shade' : True,'linewidth': 3},
                 label = airline)
    
# Plot formatting
plt.legend(prop={'size': 10}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

```python
airlines = ['United Air Lines Inc.', 'American Airlines Inc.','Alaska Airlines Inc.']

# Iterate through the five airlines
for airline in airlines:
    # Subset to the airline
    subset = df[df['name'] == airline]
    
    # Draw the density plot
    sns.distplot(subset['arr_delay'], hist = False, kde = True,
                 kde_kws = {'shade' : True,'linewidth': 3},
                 label = airline)
    
# Plot formatting
plt.legend(prop={'size': 10}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

### Since Alaska Airlines are more often earlier than United Airlines and American Airlines, but my choice would be United Airlines or American Airlines because of high density of arival on time.

```python
sample = df.arr_delay.sample(n=1000, random_state = 1)
sample.mean()
```

```python
samples_mean = []
samples_std = []
j = 0
itr = 0
while itr < 100:
    j += 1
    sample = df.arr_delay.sample(n=1000, random_state=j)
    sample_mean = sample.mean()
    sample_std = sample.std()
    samples_mean.append(sample_mean)
    samples_std.append(sample_std)
    itr += 1
print(samples_mean[:10])
print(samples_std[:10])
```

```python
sample100 = sns.distplot(samples_mean,kde=True)
```

```python
samples2_mean = []
samples2_std = []
j2 = 0
itr2 = 0
while itr2 < 10000:
    j2 += 1
    sample = df.arr_delay.sample(n=1000, random_state=j2)
    sample_mean = sample.mean()
    sample_std = sample.std()
    samples2_mean.append(sample_mean)
    samples2_std.append(sample_std)
    itr2 += 1
```

```python
sample10000 = sns.distplot(samples2_mean,kde=True)
```

### -Larger the sample size the closer would be the sample's mean to the populations mean as proved above
### -Increase in sample size leads to smaller spread in sampling distribution
